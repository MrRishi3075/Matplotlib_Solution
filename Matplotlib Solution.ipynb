### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

#### Matplotlib is a popular Python library used for creating static, animated, and interactive data visualizations in 2D and 3D. It provides a wide range of tools for creating various types of plots, charts, and graphs, making it a powerful tool for data visualization and presentation.

##### Here are some key points about Matplotlib and its usage:

1. Data Visualization: Matplotlib is primarily used for creating a wide variety of data visualizations, ranging from simple line plots to complex 3D plots.

2. Customization: It offers extensive customization options, allowing users to control the appearance of their plots, including colors, labels, fonts, and more.

3. Compatibility: Matplotlib is compatible with many different data formats and can be used in various data science and scientific computing libraries and environments, making it versatile for data analysis tasks.

4. Integration: It can be integrated with other libraries like NumPy and Pandas, making it a powerful tool for visualizing data stored in these data structures.

5. Publication-Quality Graphics: Matplotlib is often used for creating publication-quality graphics for research papers, reports, and presentations.

#### Some common types of plots that can be created using the pyplot module of Matplotlib include:

1. Line Plot: Used to visualize data points connected by straight lines. It is suitable for displaying trends over time or relationships between variables.

2. Scatter Plot: Shows individual data points as markers. It is useful for examining the distribution and relationships between two variables.

3. Bar Plot: Displays data using rectangular bars, making it suitable for comparing categories or showing data distributions for categorical variables.

4. Histogram: Represents the distribution of a single variable by dividing it into bins and displaying the frequency or density of each bin. Useful for understanding data distributions.

5. Pie Chart: Used for displaying data as a circular chart divided into slices, with each slice representing a proportion of the whole. Pie charts are effective for showing parts of a whole or percentages.

##### These are just a few examples of the types of plots you can create using Matplotlib's pyplot module. Matplotlib offers many more plot types and customization options to cater to various data visualization needs.







### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot. (Note: Also add title, xlabel, and ylabel to the plot.)

```python
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(3)

# Generate data for x and y
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, c='blue', marker='o', edgecolors='black')

# Add a title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.grid(True)
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

### The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange and display multiple plots in a grid-like fashion, which is useful for comparing and visualizing different datasets or aspects of your data in one figure.

#### Here's how you can use the subplot() function to create four line plots with the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data for the four lines
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots with a 2x2 grid (4 plots)
plt.subplot(2, 2, 1)  # Plot 1: Top left
plt.plot(x, y1, label='Line 1', color='blue')
plt.title('Line Plot 1')
plt.legend()

plt.subplot(2, 2, 2)  # Plot 2: Top right
plt.plot(x, y2, label='Line 2', color='green')
plt.title('Line Plot 2')
plt.legend()

plt.subplot(2, 2, 3)  # Plot 3: Bottom left
plt.plot(x, y3, label='Line 3', color='red')
plt.title('Line Plot 3')
plt.legend()

plt.subplot(2, 2, 4)  # Plot 4: Bottom right
plt.plot(x, y4, label='Line 4', color='purple')
plt.title('Line Plot 4')
plt.legend()

# Adjust the layout for better spacing
plt.tight_layout()

# Show the combined figure with subplots
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
```

#### A bar plot, also known as a bar chart or bar graph, is a data visualization that represents categorical data with rectangular bars. Each bar's length or height is proportional to the value it represents, making it a useful way to compare different categories or show the distribution of a single categorical variable. Bar plots can be either vertical (where bars extend vertically along the y-axis) or horizontal (where bars extend horizontally along the x-axis).

#### Bar plots are used for various purposes, including:

1. Comparison: They allow you to compare the values of different categories or groups easily.

2. Distribution: They can show the distribution of a categorical variable, such as the frequency of each category.

3. Ranking: They help in ranking categories based on their values.

4. Trends: They can reveal trends or patterns in categorical data.

##### Here's how to create both a vertical and a horizontal bar plot using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 4))  # Optional: adjust the figure size
plt.bar(company, profit, color='blue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits (Vertical Bar Plot)')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(6, 4))  # Optional: adjust the figure size
plt.barh(company, profit, color='green')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.tight_layout()
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
```

### A box plot, also known as a box-and-whisker plot, is a data visualization that displays the distribution of a dataset's values along with statistical summary information. It is used to visualize the spread, central tendency, and presence of outliers in the data. Box plots are particularly useful for comparing multiple datasets or identifying patterns in a single dataset.

#### Here's what a box plot typically represents:

1. Box: The box represents the interquartile range (IQR), which contains the middle 50% of the data. The top and bottom edges of the box correspond to the first quartile (25th percentile) and third quartile (75th percentile), respectively. The height of the box represents the spread of the middle 50% of the data.

2. Whiskers: The whiskers extend from the box and typically represent the range of the data within a certain distance from the quartiles. The exact definition of the whiskers can vary, but they often extend 1.5 times the IQR from the quartiles. Data points beyond the whiskers are considered outliers and are often plotted individually.

3. Outliers: Individual data points that fall outside the whiskers are shown as individual points or dots and are considered potential outliers.

#### Box plots are used for the following purposes:

1. Identifying Outliers: Box plots make it easy to identify potential outliers in the data by showing them as individual points.

2. Comparing Distributions: Box plots allow you to compare the distributions of multiple datasets side by side.

3. Visualizing Spread and Central Tendency: They provide a clear visual representation of both the spread (variability) and central tendency (median) of the data.

#### To create a box plot for the provided data, you can use Matplotlib. Here's how to do it:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for two boxes
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot of Two Datasets')
plt.grid(True)
plt.show()
```
